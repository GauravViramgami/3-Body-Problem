Since, the celectial orbits preserve energy, we have to use <b>Symplectic</b> methods to numerically solve the trajectories of such objects. We have majorly discussed three methods:
- Euler Method           -  `Non Energy Preserving`
- Euler Cromer Method   -  `Energy Preserving`
- Velocity Verlet Method -  `Energy Preserving`

### <center><b>Case: 1</b></center>
### <center>`2 Body System: Sun - Earth System`</center>


```python
stepsize = 0.002
num_iterations = 100000
case1 = TwoBodySystem(OBJECTS["SUN"], OBJECTS["EARTH"])
```

> 1. Non Energy Preserving Euler Method

```python
case1.euler_method(stepsize, num_iterations)
case1.plot_euler_trajectory()
```

We can observe that the object is orbiting in a spiral path since its energy is not conserved and it ends up colliding with the central object. <br>
To avoid such situation we need to use <b>Sympletic</b> methods to solve the numerical problems which involves Energey Preserving Nature.<br><br>
The following two methods: `Euler Cromer Method` and `Verlet Method` are Energy Preserving Methods, thus giving stable orbit.

```python
case1.euler_cromer_method(stepsize, num_iterations)
case1.plot_euler_cromer_trajectory()

case1.verlet_method(stepsize, num_iterations)
case1.plot_verlet_trajectory()
```
